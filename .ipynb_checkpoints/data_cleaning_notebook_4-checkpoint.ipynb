```python
age_to_fill = 32 # Starting age
start_season = 2008  # Starting season

# Loop from start_season down to 2008, decrementing the age and season each iteration
for season in range(start_season, 2007, -1):  # End at 2008
    total_years_df.loc[(total_years_df['Name'] == 'scott elarton') & (total_years_df['season'] == season), 'Age'] = age_to_fill
    age_to_fill -= 1  # Decrement the age for the next iteration

# Verify the changes for a range of seasons to see if the loop worked as expected
print(total_years_df[(total_years_df['Name'] == 'scott elarton') & (total_years_df['season'].between(2008, 2019))][['Name', 'season', 'Age']])
```
