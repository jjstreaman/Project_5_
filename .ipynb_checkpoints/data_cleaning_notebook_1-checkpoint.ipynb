```python
def convert_surgery_date_to_year(item):
    # Handle cases where the item is 0, indicating no surgery
    if item == 0:
        return 0
    # Handle non-list cases, directly converting to year if it's a valid date
    elif not isinstance(item, list):
        return item.year if pd.notnull(item) else 0
    # Handle list of dates, extracting the year from each valid date
    else:
        return [date.year for date in item if pd.notnull(date)]

# Apply the function to the column
new_merge_df['TJ Surgery Year'] = new_merge_df['TJ Surgery Date'].apply(convert_surgery_date_to_year)
```

```python
def safe_convert_tj_year(entry):
    if isinstance(entry, str):  # If the entry is a string, attempt to literal_eval it
        try:
            # Attempt to evaluate the string to a Python object (list)
            evaluated = literal_eval(entry)
            if isinstance(evaluated, list):
                return [int(i) for i in evaluated if i != 0]
            else:
                return []
        except (ValueError, SyntaxError):
            # In case of evaluation error, return an empty list
            return []
    elif isinstance(entry, list):  # If the entry is already a list, process it directly
        return [int(i) for i in entry if i != 0]
    else:
        # For any other data type, return an empty list
        return []

# Example usage on a simulated column (simulating 'TJ Surgery Year' with various possible formats)
simulated_tj_year = ['[2014, 2016]', '[0]', '[]', [2018, 2019], 0, 'not a list']

# Apply the safe conversion function
converted_tj_year = [safe_convert_tj_year(item) for item in simulated_tj_year]

# Show the result of the conversion
converted_tj_year
```

```python
# Apply the safe conversion function to the 'TJ Surgery Year' column in your actual DataFrame
new_merge_df['TJ Surgery Year'] = new_merge_df['TJ Surgery Year'].apply(safe_convert_tj_year)
```

```python
# Step 1: Determine the range of years
min_year = new_merge_df['Career Start'].min()
max_year = new_merge_df['Career End'].max()
years_range = np.arange(min_year, max_year + 1)

# Step 2: Initialize count series
active_pitchers_count = pd.Series(0, index=years_range)
tj_surgery_count = pd.Series(0, index=years_range)

# Step 3: Calculate active pitchers count
for index, row in new_merge_df.iterrows():
    active_years = set(range(row['Career Start'], row['Career End'] + 1)) - set(row['Inactive Years'])
    for year in active_years:
        if year in active_pitchers_count.index:
            active_pitchers_count.loc[year] += 1

# Step 4: Calculate TJ surgery count
for index, row in new_merge_df.iterrows():
    for surgery_year in row['TJ Surgery Year']:
        if surgery_year in tj_surgery_count.index:
            tj_surgery_count.loc[surgery_year] += 1
```

```python
min_year = active_pitchers_count.index.min()
max_year = active_pitchers_count.index.max()

fig, ax1 = plt.subplots(figsize=(12, 8))
ax1.bar(active_pitchers_count.index, active_pitchers_count.values, color='skyblue', label='Active Pitchers', width=0.4)
ax1.set_xlabel('Year')
ax1.set_ylabel('Count of Active Pitchers')
ax1.set_title('Count of Active Pitchers Over Years')
ax1.legend(loc='upper left')
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.set_xticks(np.arange(min_year, max_year + 1, 5))  # Adjusting x-axis ticks
plt.show()

fig, ax2 = plt.subplots(figsize=(12, 8))
ax2.plot(tj_surgery_count.index, tj_surgery_count.values, color='red', marker='o', linestyle='-', linewidth=2, label='TJ Surgery Count')
ax2.set_xlabel('Year')
ax2.set_ylabel('TJ Surgery Count')
ax2.set_title('TJ Surgery Count Over Years')
ax2.legend(loc='upper left')
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
ax2.set_xticks(np.arange(min_year, max_year + 1, 5))  # Adjusting x-axis ticks
plt.show()
```
