```python
all_2008_stats_df.drop(columns=['batter', 'events', 'description', 'zone',
                                'des', 'game_type', 'stand', 'home_team',
                                'away_team', 'type', 'hit_location', 'bb_type',
                                'balls', 'strikes', 'pfx_x', 'spin_dir', 
                                'pfx_z', 'plate_x', 'plate_z', 'on_3b',
                                'on_2b', 'on_1b', 'outs_when_up', 'inning',
                                'inning_topbot', 'hc_x', 'hc_y', 'fielder_2',
                                'umpire', 'sv_id', 'hit_distance_sc', 'sz_top',
                                 'sz_bot', 'launch_speed', 'launch_angle', 'game_pk',
                                'pitcher.1', 'fielder_2.1', 'fielder_3', 'fielder_4',
                                'fielder_5', 'fielder_6', 'fielder_7', 'fielder_8',
                                'fielder_9', 'estimated_ba_using_speedangle',
                                'estimated_woba_using_speedangle', 'babip_value',
                                'launch_speed_angle', 'woba_value', 'woba_denom', 'iso_value', 
                                'at_bat_number', 'pitch_number', 'home_score', 'away_score',
                                'bat_score', 'fld_score', 'post_home_score', 'post_bat_score',
                                'post_fld_score', 'post_away_score', 'if_fielding_alignment',
                                'of_fielding_alignment', 'delta_home_win_exp',
                                'delta_run_exp', 'spin_rate_deprecated', 'break_angle_deprecated',
                               'break_length_deprecated', 'tfs_deprecated', 'tfs_zulu_deprecated',
                               'spin_axis', 'effective_speed', 'release_spin_rate', 'release_extension'], inplace=True)
all_2008_stats_df.head()
```

```python
all_2008_stats_df.info()
```

```python
all_2008_stats_df = all_2008_stats_df.dropna(axis=0)
```

```python
all_2008_stats_df.reset_index(inplace=True)
```

```python
all_2008_stats_df.drop('index', axis=1)
```

```python
# Group by 'game_date' and 'pitcher' to calculate the total pitches
total_pitches = all_2008_stats_df.groupby(['game_date', 'pitcher', 'player_name']).size().reset_index(name='total_pitches')

# Group by 'game_date', 'pitcher', and 'pitch_type' to calculate the sum total of each pitch_type
total_pitches_by_type = all_2008_stats_df.groupby(['game_date', 'pitcher', 'player_name', 'pitch_type']).size().reset_index(name='count_by_pitch_type')

# Calculate averages of the specified metrics for each pitch type, grouped by game_date and pitcher
avg_metrics = all_2008_stats_df.groupby(['game_date', 'pitcher', 'player_name', 'pitch_type']).agg({
    'release_speed': 'mean',
    'release_pos_x': 'mean',
    'release_pos_z': 'mean',
    'vx0': 'mean',
    'vy0': 'mean',
    'vz0': 'mean',
    'ax': 'mean',
    'ay': 'mean',
    'az': 'mean',
    'release_pos_y': 'mean',
}).reset_index()

grouped_2008_df = total_pitches.merge(total_pitches_by_type, on=['game_date', 'pitcher', 'player_name'], how='left')
grouped_2008_df = grouped_2008_df.merge(avg_metrics, on=['game_date', 'pitcher', 'player_name', 'pitch_type'], how='left')

grouped_2008_df
```

```python
grouped_2008_df['game_date'] = pd.to_datetime(grouped_2008_df['game_date'])
grouped_2008_df['season'] = grouped_2008_df['game_date'].dt.year

# Step 1: Season Total Pitches
season_total_pitches = grouped_2008_df.groupby(['pitcher', 'player_name', 'season'])['total_pitches'].sum().reset_index(name='season_total_pitches')

# Step 2: Season Total by Pitch Type
season_total_by_pitch_type = grouped_2008_df.groupby(['pitcher', 'player_name', 'season', 'pitch_type'])['count_by_pitch_type'].sum().reset_index(name='season_total_count_by_pitch_type')

# Weighted Averages Calculation Setup
weighted_avg_columns = ['release_speed', 'release_pos_x', 'release_pos_z', 'vx0', 'vy0', 'vz0', 'ax', 'ay', 'az', 'release_pos_y']
for col in weighted_avg_columns:
    grouped_2008_df[f'{col}_product'] = grouped_2008_df[col] * grouped_2008_df['count_by_pitch_type']

weighted_avg_aggregations = {f'{col}_product': 'sum' for col in weighted_avg_columns}
weighted_avg_aggregations['count_by_pitch_type'] = 'sum'

# Aggregate for weighted averages
weighted_avg_df = grouped_2008_df.groupby(['pitcher', 'player_name', 'season', 'pitch_type']).agg(weighted_avg_aggregations).reset_index()

# Calculate weighted averages
for col in weighted_avg_columns:
    weighted_avg_df[f'{col}_weighted_avg'] = weighted_avg_df[f'{col}_product'] / weighted_avg_df['count_by_pitch_type']

# Cleanup
weighted_avg_df.drop(columns=[f'{col}_product' for col in weighted_avg_columns], inplace=True)

# Merge season totals and weighted averages
final_2008_df = pd.merge(season_total_pitches, season_total_by_pitch_type, on=['pitcher', 'player_name', 'season'])
final_2008_df = pd.merge(final_2008_df, weighted_avg_df, on=['pitcher', 'player_name', 'season', 'pitch_type'])

final_2008_df.head()
```

```python
final_2008_df['player_name'] = final_2008_df['player_name'].str.lower().str.strip().str.replace(r'\*$', '', regex=True)
final_2008_df
```

```python
print(final_2008_df['player_name'].unique())
```

```python
def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return "".join([c for c in nfkd_form if not unicodedata.combining(c)])

def clean_name(name):
    name = name.lower()
    name = remove_accents(name)
    name = re.sub(r'[-.]', '', name)
    name = re.sub(r'\s+', ' ', name).strip()
    return name

final_2008_df['player_name'] = final_2008_df['player_name'].apply(clean_name)
```

```python
# Convert 'player_name' from "last name, first name" to "first name last name"
final_2008_df['Name'] = final_2008_df['player_name'].apply(lambda x: ' '.join(x.split(', ')[::-1]))
```

Finally!

Now merge the two DF on 'Name' and 'season'.

```python
# Now, you can perform the merge using 'Name' and 'season' as the keys for both DataFrames
better_2008_df = pd.merge(final_2008_df, 
                          cleaning_filtered_df[['Name', 'season', 'Age']],
                          on=['Name', 'season'],
                          how='left')
```
